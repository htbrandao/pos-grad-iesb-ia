# # Dendrograma

Sobre Dendrograma:

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
plt.figure(figsize=(8, 5))
plt.title("Dendrograma")
dendro = dendrogram(linkage(df, method='ward'))
plt.axhline(182000, linewidth=1, color='black', linestyle='--')
plt.text(x=1500, y=183000, s='segmentgação',
    fontdict=dict(color='green',size=10),
    bbox=dict(facecolor='yellow',alpha=0.5))
plt.show()
```

# # Agglomerative Clustering

Sobre Agglomerative Clustering:

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
agg.fit(df)
```

```python
len(agg.labels_), agg.labels_
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Delicassen'], y=df['Fresh'], hue=agg.labels_)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Grocery'], y=df['Milk'], hue=agg.labels_)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Fresh'], y=df['Frozen'], hue=agg.labels_)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Detergents_Paper'], y=df['Frozen'], hue=agg.labels_)
plt.show()
```

```python
df['cluster'] = agg.labels_
```

```python
df.head()
```

```python
pairplot = sns.pairplot(df, hue='cluster')
pairplot.savefig('pairplot_clusters.png')
plt.show()
```
